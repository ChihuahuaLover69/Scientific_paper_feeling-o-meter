```python
# CÓDIGO PARA TAMBIÉN OBTENER EL AÑO DE PUBLICACIÓN DE LOS ARTÍCULOS
corpus = []

with open(jsonpath+"/arxiv-metadata-oai-snapshot.json", "r") as f: # ABRE EL ARCHIVO JSON PARA LEERLO
    parser = ijson.items(f, '', multiple_values=True) # CREA UN PARSER PARA LEER EL ARCHIVO JSON DE FORMA ITERATIVA, SIN CARGAR TODO EN MEMORIA
    c = 0
    for obj in parser:
        c+=1
        year = int(obj["update_date"][:4])
        corpus.append({
            "abstract": obj["abstract"],
            "year": year,
            "category": obj["categories"].split()[0] # OBTENEMOS LA PRIMERA CATEGORÍA DE CADA ARTÍCULO
        })
        if c>100:
            break
```

```python
df_datos = pd.DataFrame(corpus) # CREAMOS UN DATAFRAME DE PANDAS A PARTIR DE LA LISTA DE RESUMENES Y AÑOS
df_datos.head(20) # MOSTRAMOS LAS PRIMERAS FILAS DEL DATAFRAME PARA VER SU ESTRUCTURA
```

```python
from sklearn.feature_extraction.text import CountVectorizer as CV # IMPORTAMOS LA CLASE CountVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS # IMPORTAMOS LAS PALABRAS VACIAS EN INGLES
```

```python
contador.get_feature_names_out() # NOS DEVUELVE UNA LISTA CON LAS PALABRAS QUE HAN SIDO CONTADAS, ES DECIR, LAS PALABRAS QUE APARECEN EN LOS RESUMENES Y QUE NO SON PALABRAS VACIAS EN INGLES
```

```python
from nltk.corpus import sentiwordnet as swn # 

from nltk.corpus import wordnet as wn

# ESTA FUNCION RECIBE UNA PALABRA Y DEVUELVE SU SENTIMIENTO, QUE ES LA DIFERENCIA ENTRE SU PUNTUACION POSITIVA Y SU PUNTUACION NEGATIVA EN SentiWordNet. SI LA PALABRA NO TIENE UN SYNSET EN WordNet, DEVUELVE 0.
def sentimiento_palabra(palabra):
    synsets = wn.synsets(palabra)
    if not synsets:
        return 0
    
    swn_syn = swn.senti_synset(synsets[0].name())
    return swn_syn.pos_score() - swn_syn.neg_score()

sentimientos = {p: sentimiento_palabra(p) for p in df.columns}
```

#### Lo que hace la siguiente función es que toma el abstract de cada articulo lo separa por palabra con '.split()' y lo compara con su valor sentimental en el diccionario de sentimientos para calcular el sentimiento general del abstract/articulo

```python
def calcular_sentimiento_general(texto, diccionario):
    palabras = texto.split()
    if not palabras: 
        return 0
    
    # BUSCAMOS LA PALABRA EN EL DICCIONARIO DE SENTIMIENTOS, Y SI NO LA ENCONTRAMOS, DEVOLVEMOS 0. ASI, SI UNA PALABRA NO TIENE UN SENTIMIENTO ASOCIADO, NO AFECTA EL CALCULO DEL SENTIMIENTO GENERAL DEL TEXTO.
    # .get(palabra, 0) DEVUELVE 0 SI LA PALABRA NO ESTA EN EL DICCIONARIO
    puntajes = [diccionario.get(p.lower(), 0) for p in palabras]
    
    return sum(puntajes) / len(palabras)
```

```python
#Sentiwordnet es un tipo diccionario que asigna emociones a ciertas palabras, existen mas diccionarios de este tipo,
#WordNet, que se va más al significado y sinonimos
df_datos['sentimiento_neto'] = df_datos['abstract'].apply(
    lambda x: calcular_sentimiento_general(x, sentimientos)
)
```
